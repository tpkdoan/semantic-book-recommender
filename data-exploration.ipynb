```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Visualises the missing values in dataset using heatmap.

# Creates a new set of axes on which the heatmap will be plotted. ax is used to manage and customize the plot.
ax = plt.axes()
# books.isna(): Checks for missing values within the DF. It returns a boolean DF where True represents a missing value.
# .transpose(): Tranposes the DF so that rows become columns and columns become rows. This make it easier to plot missing data by column.
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing values")

plt.show()
```

Subtitle is mostly missing.
Middle variables like categories, thumbnail, description -> mostly looking random
Look at average_rating, num_pages, ratings_count -> see pattern that one of them missing, all of them missing -> these are probably from another dataset, this dataset did not contain all the books in the bigger books dataset -> maybe a source of bias because the books that are missing because they're newer or because they have other characteristics.

```python
import numpy as np

books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_books"] = 2024 - books["published_year"]
```

```python
columns_of_interest = ["num_pages", "age_of_books", "missing_description", "average_rating"]

correlation_matrix = books[columns_of_interest].corr(method="spearman")

sns.set_theme(style="white")
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={"label" : "Spearman Correlation"})

heatmap.set_title("Correlation Heatmap")

plt.show()
```

To see whether these variables have high correlation with missing_description.
-> Number of pages was correlated with missing description. It indicates that either longer books or shorter books had a missing description. As you see, we are not going to use books that have a missing description. So that our recommender may be biased towards shorter or longer book. Same as with the age of the books and average rating. And see at the value, they're not very strong -> just give som context, nothing really there.

```python
books_missing = books[~(books["description"].isna()) &
      ~(books["num_pages"].isna()) &
      ~(books["published_year"].isna()) &
      ~(books["average_rating"].isna())]

books_missing
```

```python
books_missing_25_words = books_missing[books_missing["words_in_description"] >= 25]
```

```python
books_missing_25_words["title_and_subtitle"] = (np.where(books_missing_25_words["subtitle"].isna(), books_missing_25_words["title"], books_missing_25_words[["title", "subtitle"]].astype(str).agg(" : ".join, axis=1)))
```

```python
books_missing_25_words
```

```python
books_missing_25_words["tagged_description"] = books_missing_25_words[["isbn13", "description"]].astype(str).agg("  ".join, axis=1)
```
